```python
import ta
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import pandas_datareader as pdr
```

## 1.数据准备
### 获取数据

```python
stock = pdr.get_data_yahoo('^GSPC',start = '2017-1-1', end = '2019-8-4')
stock.head()
```

### 计算指标SMA,CCI

```python
stock['sma'] = stock['Close'].rolling(5).mean()
stock['cci'] = ta.trend.cci(stock['High'],stock['Low'],stock['Close'],n = 20)
```

```python
plt.subplot(2,1,1)
plt.title('S&P index')
plt.gca().axes.get_xaxis().set_visible(False)  #这里表示的意思就是不显示横坐标轴
stock['Close'].plot(figsize = (10,8))
stock['sma'].plot(figsize = (10,8))
plt.legend()
plt.subplot(2,1,2)
stock['cci'].plot(figsize = (10,8))
plt.legend()
plt.show()
```

## 2. 交易信号、持仓信号和策略逻辑

### 交易信号
#### 产生开仓信号时应使用昨日及前日数据，以避免未来数据
#### 因为我们要算收盘价和均线的比较，所以我们要算昨天的和前天的
#### 也就是说前天的收盘价是小于均线的，但是昨天的收盘价是大于均线的，那么可能会触发一个买入信号，而买入信号是否执行，我们还要看看昨天CCI的点位是否是小于-100

```python
stock['yes_close'] = stock['Close'].shift(1)
stock['yes_sma'] = stock['sma'].shift(1)
stock['yes_cci'] = stock['cci'].shift(1)   #CCI是作为策略的一个过滤器；
stock['daybeforeyes_close'] = stock['Close'].shift(2)
stock['daybeforeyes_sma'] = stock['sma'].shift(2)
stock.head()
```

### 产生交易信号
#### sma开多信号：昨天股价上穿SMA；

```python
stock['sma_signal'] = np.where(np.logical_and(stock['daybeforeyes_close']<stock['daybeforeyes_sma'],
                                                              stock['yes_close']>stock['yes_sma']), 1, 0)
```

#### sma开空信号：昨天股价下穿SMA

```python
stock['sma_signal'] = np.where(np.logical_and(stock['daybeforeyes_close']>stock['daybeforeyes_sma'],
                                                               stock['yes_close']<stock['yes_sma']),-1,stock['sma_signal'])
```

#### 产生cci做多过滤信号

```python
stock['cci_filter'] = np.where(stock['yes_cci'] < -100, 1, 0)
```

#### 产生cci做空过滤信号

```python
stock['cci_filter']  = np.where(stock['yes_cci'] > 100,-1, stock['cci_filter'])
```

#### 过滤后的开多信号

```python
stock['filtered_signal'] = np.where(stock['sma_signal']+stock['cci_filter']==2, 1, 0)
```

#### 过滤后的开空信号

```python
stock['filtered_signal'] = np.where(stock['sma_signal']+stock['cci_filter']==-2, -1,stock['filtered_signal'])
# 生成交易信号
```

### 绘制cci，sma指标图

```python
plt.subplot(3,1,1)
plt.title('S&P CCI SMA指标图')
plt.gca().axes.get_xaxis().set_visible(False)
stock['Close'].plot(figsize=(12,8))
stock['sma'].plot()
plt.legend(loc='upper left')
plt.subplot(3,1,2)
stock['cci'].plot(figsize=(12, 8))
plt.legend(loc='upper left')
plt.subplot(3,1,3)
stock['filtered_signal'].plot(figsize=(12, 8), marker='o', linestyle='')
plt.legend(loc='upper left')
plt.show()
```

### 持仓信号

```python
# 记录持仓情况，默认为0
position = 0
# 对每一交易日进行循环
for i, item in stock.iterrows():
    # 判断交易信号
    if item['filtered_signal'] == 1:
        # 交易信号为1，则记录仓位为1
        position = 1
    elif item['filtered_signal'] == -1:
        # 交易信号为-1， 则记录仓位为-1
        position = -1
    else:
        pass
    # 记录每日持仓情况
    stock.loc[i, 'position'] = position
stock.head()
```

```python
plt.subplot(3, 1, 1)
plt.title('S&P CCI持仓图')
plt.gca().axes.get_xaxis().set_visible(False)
stock['Close'].plot(figsize = (12,12))
plt.legend()
plt.subplot(3, 1, 2)
stock['cci'].plot(figsize = (12,12))
plt.legend()
plt.gca().axes.get_xaxis().set_visible(False)
plt.subplot(3, 1, 3)
stock['position'].plot(marker='o', figsize=(12,12),linestyle='')
plt.legend()
plt.show()
```

## 3.策略收益和数据可视化

### 计算策略收益

#### 计算股票每日收益率

```python
stock['pct_change'] = stock['Close'].pct_change()
```

#### 计算策略每日收益率

```python
stock['strategy_return'] = stock['pct_change'] * stock['position']
```

#### 计算股票累积收益率

```python
stock['return'] = (stock['pct_change']+1).cumprod()
```

#### 计算策略累积收益率

```python
stock['strategy_cum_return'] = (1 + stock['strategy_return']).cumprod()
stock.head()
```

### 将股票累积收益率和策略累积收益率绘图

```python
stock[['return', 'strategy_cum_return']].plot(figsize = (12,6))

plt.title('S&P CCI收益图')
plt.legend()
plt.show()
```
