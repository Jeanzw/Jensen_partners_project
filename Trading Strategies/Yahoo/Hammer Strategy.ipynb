## 1. 收集并计算所需数据

```python
import numpy as np
import pandas as pd
import pandas_datareader as pdr
```

```python
#我们为了让代码好让别人看懂，最好在最开始就把我们设定的参数都写清楚
stock = 'TCEHY'           #股票代码，腾讯股票
body_size = 0.03          #表示锤子实体大小上限，基准为当日开盘价，实体不能太大，波动范围限制在3%之间
head_size = 0.5           #表示锤子上影线长度上限，基准为下引线长度，上影线要短一点，不能超过下引线的一半
tail_size = 2             #表示下影线与实体大小比值，下影线要大于实体两倍
length = 10               #表示观察期时间长短
stoplose_trigger = 1      #表示当价格偏高均线满足几倍标准差时要止损。移动止损（动态）：mean - 标准差
```

```python
data = pdr.get_data_yahoo(stock, start = '2015-1-1', end = '2019-8-2')
data.head()
```

```python
data.reset_index(inplace=True)        #把索引设置成为默认；为了后面交易策略逻辑循环更方便一些；

data['pct_change'] = data['Close'].pct_change()
#data['pct_change'] = data['Close']/data['Close'].shift(1) - 1   ->  也可以写成这样
data.head()
```

```python
data['ma'] = data['Close'].rolling(length).mean()
data['std'] = data['Close'].rolling(length).std()
```

```python
#将我们不要的列给删除
#这里还可以用drop：https://blog.csdn.net/nuaadot/article/details/78304642
del data['Adj Close']  
del data['Volume']
data.head()
```

```python
#由于实盘中当天的日线级别参考指标未实现，因此使用昨日参考指标指导当日交易，避免未来函数
data['yes_ma'] = data['ma'].shift(1)
data['yes_std'] = data['std'].shift(1)
```

## 2. 识别锤子形态

### 计算实体，上影线和下影线

```python
data['body'] = abs(data['Close'] - data['Open'])
data['head'] = data['High'] - data[['Open','Close']].max(axis = 1)  #计算上影线，axis =1说明我们是一行行去计算，而不是纵向去计算，如果不加axis = 1那么就相当于我们在所有数据中的close和open中选择最高和最低
data['tail'] = data[['Open','Close']].max(axis = 1) - data['Low']   #计算下影线
data.head()
```

### 判断K线部分是否符合锤子线要求

```python
data['body_cond'] = np.where(data['body']/data['Open'] < body_size, 1, 0)  #帮我们判断实体的大小是否符合前面的条件
data['head_cond'] = np.where(data['tail'] == 0, False, data['head']/data['tail'] < head_size)  #上影线不能超过下影线的一半
#但是为什么我们head_cond不按照body_cond一样干呢？那是因为，如果我们今天收盘价就是最低价呢？那么data['tail']就是0，那么就会报错
data['tail_cond'] = np.where(data['body'] == 0, True, data['tail']/data['body'] > tail_size)  #下影线要比实体的两倍要长
#而这里data['body'] == 0返回True，这是因为如果今天开盘价和收盘价一样，那么怎么样都是锤子线，因为已经满足了前面的几个条件了
```

### 判断K线形态是否符合锤子线

```python
data['hammer'] = data[['head_cond','body_cond','tail_cond']].all(axis = 1)  #.all()就是说要同时满足以上三个条件才是锤子K线
#.any()只要三个中有一个是true那么就是true
data[data['hammer']].tail(5)  #如果直接print(data['hammer'])那么其实就会把hammer的结果弄出来，但是如果我们采取现在这种做法就会把hammer是True给print出来
#把hammer是True的给print出来，那么出来的结果就都是锤子线
```

```python
#由于实盘中当天的日线级别参考指标未实现，因此我们应根据昨日是否满足锤子形态要求作为开仓信号
data['yes_hammer'] = data['hammer'].shift(1)
```

## 3. 编写交易逻辑——循环法

回顾一下逻辑：我们整个锤子线要在下跌情况下才有用，而这个锤子线的判断我们已经写好了
而我们的止损是有两种方式的：

①移动止损：过去十天的均值 - 一倍的标准差，如果跌落这个价格我们就移动止损
②固定止损：开仓价 - 开仓时这一天的一倍的标准差

```python
flag = 0  #持仓记录，1代表有仓位，0代表空仓
for i in range(2*length,len(data)):   #从20天开始算，因为前期数据无效。其实从10天开始就可以，10天之前就是NaN
    #如果已持仓,判断是否止损
    if flag == 1:   #这里就是说今天是有持仓的情况，就说明之前是买过股票的，所以今天才会有持仓现象
        #有持仓的情况，先计算一下今天的止损价，我们这里为了更加稳健，就移动止损和固定止损取其高
        stoplose_price = max(data.loc[i,'yes_ma'] - stoplose_trigger * data.loc[i,'yes_std'],
                             long_open_price - long_open_delta)
        #逗号前面是移动止损，就是我今天看过去十天的均价减去一倍的标准差，一倍就是stoplose_trigger就是我们之前设定的
        
        #当天价格低于止损价，则进行止损，一个是移动止损，一个是开仓时候的开仓和开仓价 - 1倍标准差
        if data.loc[i,'Low'] < stoplose_price:  #接下来要做的都是止损的操作
            flag = 0   #如果当天触发了止损条件，那么我们就要把这支股票给买了，那么就让flag = 0
            
            #data.loc[i,'return'] = stoplose_price/data.loc[i-1,'Close'] - 1  #-> 这种写法简单但不精确，因为stop_price我们可能拿不到，因为可能今天开盘价格高于了我们的止损价，然后导致我们拿不到我们的止损价
            #所以我们要用我们今天的止损价格和开盘价格取较低的数，确保一定低于止损价格
            data.loc[i, 'return'] = min(data.loc[i, 'Open'], stoplose_price)/data.loc[i-1, 'Close'] - 1 
            #计算清盘当天的收益；取min是因为，如果当天开盘价就小于了止损价，那么我们就要以开盘价就止损；
            #不然会导致策略收益高估；
            #收益计算时要除以前一天的收盘价；
            data.loc[i,'trade_mark'] = -10   #表示当天持仓并进行平仓，记录自己当天交易操作，平仓：-10
            #开仓是10；持仓为1，方便查阅
        
        #如果不满足止损条件，则继续持仓
            #继续持仓说明我们可以获得这只股票当天的收益，那么就用当天的收盘价去除以前一天的收盘价 - 1
        else:
            data.loc[i,'return'] = data.loc[i,'Close']/data.loc[i - 1,'Close'] - 1
            data.loc[i,'trade_mark'] = 1   #表示当天持仓
    
    
    
    #如果未持仓，判断是否进行开仓
    else:  #当天没有持仓的情况,如果不持仓那么直接每一天的情况走一遍看看是不是锤子线
        #判断是否为下降趋势，平均重心是下降的，锤子线开仓要满足形态和下降趋势
        if data.loc[i - length, 'yes_ma'] > data.loc[i, 'yes_ma']:  #其实这一行代码就可以说明为什么我们前面要空出20个for i in range(2*length,len(data))，因为前十个数据绝对是空值，而我们算移动止损是用十天的均值，所以至少要空出20天的数据
            #如果满足下降趋势那么接下来判断是否符合锤子形态
            if data.loc[i, 'yes_hammer']:  #我们发出锤子线的信号是不能是今天的，不然就是未来函数了。应该是昨天发出了锤子线的信号，然后等到下一天我们才能买入这只股票，这也就是为什么我们之前要用到yes_hammer这个原因
                #更改持仓标记
                flag = 1
                #记录开仓时开仓价格及标准差：是为了做固定止损
                long_open_price = data.loc[i, 'Open']
                long_open_delta = data.loc[i, 'yes_std']
                #计算当天收益率
                data.loc[i,'return'] = data.loc[i, 'Close']/data.loc[i, 'Open'] - 1  #以产生信号之后的第二天开盘价开仓，我们假设所有的股票都是按照第二天开盘价去买的
                data.loc[i,'trade_mark'] = 10   #表示当天开仓
                #当天开仓之后不进行平仓判断
```

## 4.计算策略收益率

```python
data['return'].fillna(0, inplace=True)                   #对大循环中未处理的：既没有持仓，也不满足开仓条件的日期进行处理，则让这些天的return都等于0；
data['strategy_return'] = (data['return'] + 1).cumprod()
data['stock_return'] = (data['pct_change'] + 1).cumprod()

data.tail(50)
```
