```python
import ta
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import pandas_datareader as pdr
```

## 1.数据准备

```python
stock = pdr.get_data_yahoo('^GSPC',start = '2018-7-1', end = '2019-7-29')
# https://github.com/bukosabino/ta
# https://technical-analysis-library-in-python.readthedocs.io/en/latest/
```

### 计算bolling线

```python
stock['upper'] = ta.volatility.bollinger_hband(stock['Close'],n=20, ndev=2)
stock['middle'] = ta.volatility.bollinger_mavg(stock['Close'],n=20)
stock['lower'] = ta.volatility.bollinger_lband(stock['Close'],n=20, ndev=2)
```

### 绘图

```python
fig = plt.figure(figsize = (10,6))
plt.plot(stock['Close'])
plt.plot(stock['upper'],linestyle = '-')
plt.plot(stock['middle'],linestyle = '--')
plt.plot(stock['lower'],linestyle = '-')
plt.title('S&P 500 Bollinger Band')
plt.legend()
plt.show()
```

## 2.交易信号和持仓信号

#### 交易信号和持仓信号计算（因为我们在这里用的是循环的方法，所以交易信号和持仓信号都是分开计算的）

### 计算昨天数据

```python
stock['yes_close'] = stock['Close'].shift(1)
stock['yes_lower'] = stock['lower'].shift(1)
stock['yes_upper'] = stock['upper'].shift(1)
```

### 计算前天数据

```python
stock['daybeforeyes_close'] = stock['Close'].shift(2)
stock['daybeforeyes_lower'] = stock['lower'].shift(2)
stock['daybeforeyes_upper'] = stock['lower'].shift(2)
```

```python
stock.head()
```

### 计算交易信号

#### 开多信号：前天收盘价低于下轨，而昨天收盘价高于下轨

```python
stock['signal'] = np.where(np.logical_and(stock['daybeforeyes_close'] < stock['daybeforeyes_lower'],stock['yes_close'] > stock['yes_lower']),1,0)
```

#### 开空信号：前天收盘价高于上轨，而昨天收盘价低于下轨

```python
stock['signal'] = np.where(np.logical_and(stock['daybeforeyes_close'] > stock['daybeforeyes_upper'],stock['yes_close'] < stock['yes_upper']),-1,0)
```

### 绘制交易信号图

```python
plt.subplot(2,1,1)
plt.title('S&P bollinger trading signal')
plt.gca().axes.get_xaxis().set_visible(False)
stock['Close'].plot(figsize = (10,8))
plt.plot(stock['upper'],linestyle = '-')
plt.plot(stock['middle'],linestyle = '--')
plt.plot(stock['lower'],linestyle = '-')
plt.legend()
plt.subplot(2,1,2)
plt.plot(stock['signal'],marker = 'o',linestyle = '')
plt.legend()
plt.suptitle('S&P bollinger trading signal')
plt.show()
```

#### *****  iterrows 的用法    *******

```python
df = pd.DataFrame(np.random.randn(4,2))
df.columns = ['A','B']
df.index = ['a','b','c','d']
df
```

```python
for i in df.iterrows():
    print(i)   #这样子取出来的就是每一行的内容
```

```python
for i in df.iterrows():
    print(i[1])  #这样子就是不取index，取AB组成的series
```

```python
for i in df.iterrows():
    print(i[1]['B'])     #这里我相当于是不取index，取AB组成的series，然后再在里面选取B列
```

##### 从上面例子我们可以看出其实iterrows返回的是两个东西，一个是行标签，另一个就是series，而之前的列就变成了我们这里的行
##### 比如原先有10列，那么用iterrows后返回的就是10行。而如果我们只有一个变量，比如说上面的i，那么其实返回的就是一个tuple，那么接下来我们可以进行改进

```python
#更好更容易看懂的写法
for i, item in df.iterrows():
    print(i)  #print出来的是行标签
    print(item)  #输出的series
    #这样子做就不像是前面那样子，显示出现tuple，然后再在tuple里面选取行标签和series
```

### 持仓信号
#### 上面我们只是写好了交易信号，但是持仓信号还是没有写好
#### 使用position标记持仓情况，全新的循环法思路（对于比较复杂的策略这个非常好用）

```python
position = 0   #position = 0的时候不持仓，position = 1的时候持多仓，position = -1的时候持空仓
```

#### 对每个交易日进行循环

```python
for item in stock.iterrows():  
    #python里使用iterrows()对dataframe进行遍历，返回的item其实是一个元组，元组由两部分组成，一个是行标签，一个是series
    #判断交易信号
    if item[1]['signal'] ==1:
        #交易信号为1，则记录仓位为1,说明这里可以持有多仓了
        position = 1
    elif item[1]['signal'] == -1:
        #交易信号为-1，则记录仓位为-1，说明这里持有空仓了
        position = -1
    else:
        pass  #啥都不做
    #记录每日持仓情况，for循环的核心关键
    stock.loc[item[0],'position'] = position    #如果既不是1也不是-1那么就沿用上一个position。自动往下填充的就是上一个产生的交易信号
stock.head()
```

## 3.计算策略收益及可视化

### 计算策略收益
#### 计算股票每日收益率

```python
stock['pct_change'] = stock['Close'].pct_change()
```

#### 计算股票累积收益率

```python
stock['return'] = (stock['pct_change']+1).cumprod()
```

#### 计算策略每日收益率

```python
stock['strategy_return'] = stock['position'] * stock['pct_change']
```

#### 计算策略累积收益率

```python
stock['strategy_cum_return'] = (1 + stock['strategy_return']).cumprod()
```

### 将股票累积收益率和策略累计收益率绘图

```python
stock[['return','strategy_cum_return']].plot(figsize = (10,8))
plt.title('Bollinger Bands return')
plt.legend(loc = 'upper left')
plt.show()
```
