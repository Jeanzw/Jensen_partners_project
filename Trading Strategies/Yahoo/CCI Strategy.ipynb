#### 之前我们写的策略其实交易信号&持仓是合并的，但是如果不是我们自己看，是别人来看或者是初学者来看的话，那么交易信号&持仓分开来写会更容易理解，结构更清晰

```python
import ta
# https://www.zhihu.com/question/39951384
# https://github.com/bukosabino/ta
# https://technical-analysis-library-in-python.readthedocs.io/en/latest/
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import pandas_datareader as pdr
```

## 1.数据获取和数据处理(我们这一次来跑高盛)

```python
stock = pdr.get_data_yahoo('GS',start = '2018-7-1', end = '2019-7-29')
stock.head()
```

### 计算CCI，取时间周期为20

```python
stock['cci'] = ta.trend.cci(stock['High'],stock['Low'],stock['Close'],n = 20)
stock.head()
```

### 绘制CCI指标图

```python
plt.subplot(2,1,1)
plt.title('The Goldman Sachs Group index')
plt.gca().axes.get_xaxis().set_visible(False)  #这里表示的意思就是不显示横坐标轴
stock['Close'].plot(figsize = (10,8))
plt.legend()
plt.subplot(2,1,2)
stock['cci'].plot(figsize = (10,8))
plt.axhline(100)
plt.axhline(-100)
plt.legend()
plt.show()
```

## 2.交易信号、持仓信号和策略逻辑

### 产生开仓信号时应使用昨日及前日的CCI，以避免未来数据

```python
stock['yes_cci'] = stock['cci'].shift(1)
stock['daybeforeyes_cci'] = stock['cci'].shift(2)
stock
```

### 产生开仓信号
#### 开多信号，当前天CCI小于-100，昨日CCI大于-100则记为开多信号

```python
stock['signal'] = np.where(np.logical_and(stock['daybeforeyes_cci'] < -100,stock['yes_cci']>-100),1,np.nan)  
#这里我们用了一个逻辑语，然后要让前面两个同时成立才是1
```

#### 开空信号，当前天CCI大于100，昨日CCI小于100则记为开空信号

```python
stock['signal'] = np.where(np.logical_and(stock['daybeforeyes_cci'] > 100,stock['yes_cci'] < 100),-1,stock['signal'])
```

```python
stock['signal'] = stock['signal'].fillna(method = 'ffill')   #在下一个信号产生之前都是沿用下一个信号
stock['signal'] = stock['signal'].fillna(0)   #刚开始期间其实也没有任何的信号，所以就是空值，所以相当于我们把最前面一段的空值给填充
stock.head()
```

```python
plt.subplot(3,1,1)
plt.title('open a position')
plt.gca().axes.get_xaxis().set_visible(False)
stock['Close'].plot(figsize = (10,8))
plt.legend(loc = 'upper left')
plt.subplot(3,1,2)
stock['cci'].plot(figsize = (10,8))
plt.gca().axes.get_xaxis().set_visible(False)
plt.legend(loc = 'upper left')
plt.subplot(3,1,3)
stock['signal'].plot(figsize = (10,8),marker = 'o',linestyle = '')
plt.legend(loc = 'upper left')
plt.show()
```

## 3.收益计算和净值绘制

#### 计算策略收益
#### 计算股票每日收益率

```python
stock['pct_change'] = stock['Close'].pct_change()
```

#### 计算策略每日收益率

```python
stock['strategy_return'] = stock['pct_change'] * stock['signal'].shift(1)
```

#### 计算股票累积收益率

```python
stock['return'] = (stock['pct_change']+1).cumprod()
```

#### 计算策略累积收益率

```python
stock['strategy_cum_return'] = (1 + stock['strategy_return']).cumprod()
```

#### 将股票累积收益率和策略累计收益率绘图

```python
stock[['return','strategy_cum_return']].plot(figsize = (10,8))
plt.title('Goldman Sachs CCI return')
plt.legend(loc = 'upper left')
plt.show()
```
