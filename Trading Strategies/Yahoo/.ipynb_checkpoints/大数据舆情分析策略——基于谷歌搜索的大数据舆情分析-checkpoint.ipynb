```python
#参考论文：https://www.nature.com/articles/srep01684
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import datetime as dt
import quandl
```

## 1.数据整理

### 读取论文数据

```python
paper = pd.read_csv('paper_data.csv',sep = ' ',parse_dates = True) #注意：sep = ''这样是会报错的，除非里面有一个空格，这样子python才知道我们是按照空格来进行划分的
paper.head()  #这其实有非常多的列，但是我们需要的列其实并没有这么多,我们需要的其实就是Google End Date, Debt, Date, DJClose
```

### 我们把这四列筛选出来，然后用字典的方式进行重命名。通过字典的方式去构建dataframe

```python
#这里的Google_week是谷歌的，而Date则是道琼斯的收盘价的时间
#这里的debt是经过处理之后的，而不是说搜索人数就这么多
data = pd.DataFrame({'Google_week': paper['Google End Date'],
                     'Debt': paper['debt'].astype(np.float64),
                     'Date': paper['DJIA Date'],
                     'DJClose': paper['DJIA Closing Price'].astype(np.float64)})
data.head()
```

```python
#我们前三周的数据是不能用的，我们是最早最早都是拿第四周的数据跟前三周去进行对比
data['Date'] = pd.to_datetime(data['Date'])
data['Google_week'] = pd.to_datetime(data['Google_week'])
data.head()
```

### 读取我们自己下载的谷歌搜索指数数据

```python
#注意：论文数据作者有自己做过处理，normalize，虽然和我们自己下载的数据有非常高的相关性，但是仍然有差别
trends = pd.read_csv('debt_google_trend.csv')
trends.head()  
#这里debt对应的数字并不是真正这一周debt搜索的次数，google只会告诉我们高了还是低了，具体这个值代表的意思我们是不知道的，我们只是知道相对其他周这个数字是多了还是少了
```

### 接下来我们要把week这一列保留最后面的日期，前面的日期不要

```python
trends['Week'] = trends['Week'].apply(lambda x:pd.to_datetime(x.split(' ')[-1]))  
#x.split()可以将字符串分割成一个个小的,，而这里我们用空格来split
trends.head()
```

```python
#其实在这里我们要把道琼斯指数的debt也给下载下来
data.head()
#我们要把data以及我们上面处理后的trends给合并起来，我们用merge，然后确保数据长度一样，而且Google_week以及我们自己下载的Week是一样的
```

```python
alldata = pd.merge(data,trends,left_on = 'Google_week', right_on = 'Week')  
#如果合并的参考系是一样的名字，那么这里可以直接默认不写，但是这里数据虽然一样，但是列名不一样
alldata.head()
```

### 接下来我们要把Week给删除并且把Date设定为index

```python
#del alldata['Week']  可以直接用del,但是我们这里介绍另一种方法drop
alldata.drop('Week', inplace = True, axis = 1)  #inplace = True说明是实实在在的删除而不是在视图里面删除，axis = 1告诉我们是在列的索引上删除，如果是axis = 0那么就告诉python是在行的索引上删除
alldata.set_index('Date',inplace = True)
alldata.head()
```

```python
#现在我们其实是有两个debt的，Debt是作者论文里面的，而debt则是我们自己下载的
alldata.rename(columns = {'Debt':'Debt_paper','debt':'Debt_download'},inplace = True)
alldata.head()
```

```python
#现在我们想看看Debt_paper和Debt_download之间的相关性如何，如果相关性很强，那么我们就认为的确就是做了中性化，那么这两列数据其实没什么差别
both_trends = alldata[['Google_week','Debt_paper','Debt_download']].set_index('Google_week')
both_trends.head()
```

```python
both_trends.corr()  #我们发现相关性非常高，这其实就说明作者的数据是经过normalized的，跟我们再google trend数据库里面下载的数据是几乎一致的
```

## 2.交易信号和信号逻辑

```python
#我们策略的思路就是，先看前三周的搜索量的平均，然后用第四周进行对比，然后在第五周决定是做多还是做空
alldata.head()  #这个时候的index是Date
```

```python
alldata = alldata.reset_index().set_index('Google_week')  #我们重新把Google week当做index
alldata.head()
```

```python
alldata['MA_p'] = alldata['Debt_paper'].shift(1).rolling(window = 3).mean()  #我们一定要shift(1),因为我们策略是在第四周和前三周比较搜索量，而如果没有shift(1)那么就相当于是说我们在第三周向前滚动三周，然后用第三周和包括第三周的前三周的均值进行比较
alldata['MA_d'] = alldata['Debt_download'].shift(1).rolling(window = 3).mean()
alldata.head()
#这里的MA_p和MA_d其实都是前三周的平均。比如说2004-1-31，第四周debt_download是63，但是前三周的平均只有61，那么我们就可以进行对比，然后产生交易信号
```

### 策略产生信号

```python
alldata['signal_p'] = np.where(alldata['Debt_paper'] > alldata['MA_p'], -1, 1)  #因为debt是不好的词，所以如果第四周比前三周多，那么就要做空
alldata['signal_d'] = np.where(alldata['Debt_download'] > alldata['MA_d'], -1, 1)
alldata.head()  #我们发现signal_p和signal_d基本一样，但是有些时候paper是做了一些处理，所以最后paper的数据是要比我们自己下载的数据要好的
```

```python
#而我们另外发现前三周竟然也有信号，那是因为我们产生信号的语句的问题
#按理来说前三周应该是没有交易型号的，所以我们要把前三周的交易信号清除一下
alldata.loc[:3,['signal_p','signal_d']] = 0
alldata.head()
```

## 3.计算策略收益并可视化

```python
alldata['pct_change'] = alldata['DJClose'].pct_change()
alldata.head()
```

```python
alldata['ret_p'] = alldata['pct_change']*alldata['signal_p'].shift(1)  #我们之所以要shift(1)那是因为我们是第五周才进行买入或者卖出。所以是第四周产生交易信号，而第五周产生收益，两个相乘，所以要shift(1)
alldata['ret_d'] = alldata['pct_change']*alldata['signal_d'].shift(1)
alldata.head()  #我们会发现前四周都是没有收益的，这个就符合我们策略的情况，第五周才开始交易，那么第五周才开始会有收益
```

### 计算累计收益

```python
alldata['cumret_p'] = (1 + alldata['ret_p']).cumprod()
alldata['cumret_d'] = (1 + alldata['ret_d']).cumprod()
alldata[['cumret_p','cumret_d']].head(10)
```

### 绘图

```python
alldata[['cumret_p', 'cumret_d']].plot(figsize=(12,6));
plt.show()
```
